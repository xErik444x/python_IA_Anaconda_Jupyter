```python
import pandas as pd
import numpy as np
import matplotlib as pl
```

```python
pagina_web = "https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_y_territorios_dependientes_por_poblaci%C3%B3n"
```

```python
datos = pd.read_html(pagina_web)
datos[0]
```

```python


dataframeCensos = pd.DataFrame(datos[0]["Censo másreciente,últimaestimaciónoficial,proyecciónde la ONUo reloj depoblaciónnacional"])
dataframePaises = pd.DataFrame(datos[0]["País (o territorio dependiente)"])
dataframeDatos = pd.concat([dataframePaises,dataframeCensos],axis=1)
dataframeDatos
```

```python
dict = {'País (o territorio dependiente)': 'Pais',
        'Censo másreciente,últimaestimaciónoficial,proyecciónde la ONUo reloj depoblaciónnacional': 'Censo'}
dataframeDatos.rename(columns=dict,inplace=True)
```
