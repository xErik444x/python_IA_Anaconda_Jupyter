# Graficos con Panda

```python
import numpy as np
import pandas as pd
```

```python
                            #enteros aleatorios de 0 a 200 y 50 valores x 4 columnas
dataframe = pd.DataFrame(np.random.randint(200,size=(50,4)),columns=["a","b","c","d"])
```

```python
dataframe
```

```python
dataframe["a"].hist()
```

```python
dataframe["b"].hist(bins=30) #bins = columnas
```

```python
dataframe["a"].plot.hist()
```

```python
dataframe.plot.area()
```

```python
dataframe.plot.area(alpha=0.3)
```

```python
dataframe.plot.bar()
```

```python
dataframe.plot.bar(stacked=True)
```

```python
dataframe.plot.scatter(x="a",y="b",c="c")
```

```python
dataframe.plot.scatter(x="a",y="b",c="c",cmap="coolwarm")
```

```python
dataframe.plot.box()
```

```python
dataframe.plot.hexbin(x="a",y="b",gridsize=12,cmap="summer")
```

```python
dataframe.plot.kde()
```

```python
dataframe.plot.density()
```
