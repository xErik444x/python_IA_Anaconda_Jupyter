# Graficos de regresión (Regression Plots)

```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head()
```

```python
#grafico de regresión
sns.lmplot(x="total_bill",y="tip",data=propinas)
```

```python
sns.lmplot(x="total_bill",y="tip",data=propinas,hue="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=propinas,hue="sex",markers=["o","v"])
```

```python
sns.lmplot(x="total_bill",y="tip",data=propinas,hue="sex",markers=["o","v"],scatter_kws={"s":80})
                                                                            #scatter_kws es para cambiar el tamaño
```

```python
sns.lmplot(x="total_bill",y="tip",data=propinas,col="day",hue="sex",row="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=propinas,aspect=1.2,height=10) #aspect ancho, height altura
```
