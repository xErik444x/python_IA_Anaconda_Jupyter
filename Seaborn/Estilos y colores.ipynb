# Estilos y colores

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
propinas = sns.load_dataset("tips")
propinas.head()
```

```python
sns.countplot(x="sex",data=propinas)
```

```python
sns.set_style("whitegrid")
sns.countplot(x="sex",data=propinas)
```

```python
sns.set_style("whitegrid")
sns.countplot(x="sex",data=propinas)
sns.despine() #quitar barras laterales
```

```python
plt.figure(figsize=(12,4)) #ancho,alto
sns.set_style("white")
sns.countplot(x="sex",data=propinas)
sns.despine(left=True,bottom=True) #quitar barras 
```
