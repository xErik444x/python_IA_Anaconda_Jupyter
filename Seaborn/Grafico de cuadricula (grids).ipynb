# Grafico de cuadricula (grids) : PairGrid y FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flores = sns.load_dataset("iris")
flores.head()
```

```python
sns.pairplot(flores) #graficos de toda las columnas numericas
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter) #poner todos los graficos a tipo scatter
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot) #la diagonal de tipo distplot(montaitas)
graficos.map_upper(sns.scatterplot) 
graficos.map_lower(sns.barplot) 
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.histplot) #la diagonal de tipo distplot(montaitas)
graficos.map_upper(sns.scatterplot) 
graficos.map_lower(sns.kdeplot) 
```
