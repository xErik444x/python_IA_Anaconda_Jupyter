# Grafico de cuadricula (grids) : PairGrid y FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flores = sns.load_dataset("iris")
flores.head()
```

```python
sns.pairplot(flores) #graficos de toda las columnas numericas
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter) #poner todos los graficos a tipo scatter
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot) #la diagonal de tipo distplot(montañitas)
graficos.map_upper(sns.scatterplot) 
graficos.map_lower(sns.barplot) 
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.histplot) #la diagonal de tipo distplot(montañitas)
graficos.map_upper(sns.scatterplot) 
graficos.map_lower(sns.kdeplot) 
```

# FacetGrid

```python
propinas = sns.load_dataset("tips")
propinas.head()
```

```python
#valores unicos
propinas["time"].unique()
```

```python
propinas["day"].unique()
```

```python
graficos2 = sns.FacetGrid(data=propinas,col="time",row="smoker")
graficos2.map(sns.histplot,"total_bill")
```

```python
graficos2 = sns.FacetGrid(data=propinas,col="time",row="smoker")
graficos2.map(sns.scatterplot,"total_bill","tip") #el scatter necesita 2 parametros
```
