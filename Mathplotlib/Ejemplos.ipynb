# Ejemplos matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y,"red")
plt.title("Grafico 1")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
plt.subplot(1,2,1) #1 fila y 2 columnas/graficos y el 3er parametro es el grafico actual (posición)
plt.plot(x,y,"green")
plt.title("Subnormalidad de NEI")
plt.subplot(1,2,2)
plt.title("Subnormalidad de KRIS")
plt.plot(x,y,"red")
plt.show()
```

## Programacion orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8]) #espacio izquierda, altura y luego ancho y alto, todo con porcentaje.
grafico.plot(x,-y,"purple")
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])  
grafico1.plot(x,y,"green")
grafico2 = figura.add_axes([0.2,0.6,0.3,0.3])
grafico2.plot(x,y,"blue")
```

# Multigrafo

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)

graficos[0][0].plot(x,y,"blue")
graficos[0][1].plot(x,y,"red")
graficos[-1][-1].plot(x,y,"green")

graficos[0][0].set_title("Que se yo")
```

# tamaño del grafico

```python
figura = plt.figure(figsize=(8,4)) #figsize = alto x ancho
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,"r")
```

# Dos graficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9]) #add axes especifica el tamaño
graficos.plot(x,x**2,label="X al cuadrado")
graficos.plot(x,y**2,label="Y al cuadrado")
graficos.plot(x,x**3,label="X al cubo")
graficos.legend(loc=0) #loc 0 es la posicion optima (que no molesta)
```

# Linea del grafico

```python
figura2 = plt.figure()
grafico1 = figura2.add_axes([0,0,1,1])  
grafico1.plot(x,y,color="green",linewidth=10,alpha=0.3,linestyle=":")
```
